```python
df = pd.read_csv("confirmed_diff.csv")
```

```python
region = "Wilson - North Carolina"
region2 = "Orange - North Carolina"
subset = df[df['region'] == region]
subset2 = df[df['region'] == region2]

plt.plot(subset.Date, subset['diff'])
plt.plot(subset2.Date, subset2['diff'])
```

$\sigma = \frac{1}{1+e^{-x}}$
<br >
$1+e^{-x} = \frac{1}{\sigma}$
<br >
$ e^{-x}  = \frac{1}{\sigma} - 1$ 
<br >
$ x = - \log (\frac{1}{\sigma} - 1)$

```python
simple_lstm_model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(8, input_shape=subset["Value"].shape),
    tf.keras.layers.Dense(3)
])

simple_lstm_model.compile(optimizer='adam', loss='mae')
```
